Let us explore more about the five important features in more detail. This can be made convinient through plotting pairplot in Seaborn. I introduced regression line to see how deep the correlaion is.

```python
plt.figure(figsize=(20,6))
sns.set_style('white')
sns.set_style('ticks')
sns.pairplot(house_prices_new_numerical_topten, kind='reg')
```

From the graph above OverallQual is positively correlated with GrLivArea,GarageCars, GarageArea and TotalBsmtSF. We can clearly see that most of the top ten numerical features have strong correlation with the others.    

**A heat map can be more interactive. Let us plot a heat map between the correlated variables**

```python
corrmatrix = house_prices_new_numerical.corr()
k = 11 #number of variables for heatmap
cols = corrmatrix.nlargest(k, 'SalePrice')['SalePrice'].index
cm = np.corrcoef(house_prices_new_numerical[cols].values.T)
sns.set(font_scale=1.25)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=cols.values, 
                 xticklabels=cols.values)
plt.show()
```

- Ah! Heatmap is the best way to get a quick overview of the relationships between varaibles. The correlation between the variables is so strong that it can indicate a situation of multicollinearity.
- It is evident that there exists a strong relationship between **'GarageCars' and 'GarageArea', 'Total Basement square footage' and '1st floor square footage', 'Above grade(ground) area' and 'Total no. of rooms above grade(ground)**.

A few other noticeable correlations are between:
- OverallQual and Above ground area, Garage cars, Garage area, Year built
- Above ground area and 1st floor, Full bath
- Garage cars and OverallQual, Year built
- Year built and Year remodeled
